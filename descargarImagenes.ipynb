# Descarga de imágenes
Este notebook es para descargar las imágenes de los apartamentos de airbnb.
## Carga dataset completo
Se carga en un dataframe de Pandas el dataset de Airbnb completo

```python
import pandas as pd
import os

# Carga el dataset y devuelve un dataframe de Pandas
def load_airbnb_dataset(ruta,nombre):
    csv_path = os.path.join(ruta, nombre)
    return pd.read_csv(csv_path, sep=';')
```

### Se busca en qué columna se encuentran las url de las imágenes
Compruebo en qué columna se encuentra la url de la imagen del apartamento

```python
housing = load_airbnb_dataset('datasets','airbnb-listings.csv')

pd.options.display.max_columns = None
housing.head()
```

```python
pd.options.display.max_colwidth = 999
print(housing["Thumbnail Url"])
pd.options.display.max_colwidth = 100
```

Se observa cuántos tienen valores nulos

```python
housing["Thumbnail Url"].isnull().values.sum()
```

Se elimina las filas que contengan nulos en la url de la imagen

```python
housinNoNulaUrl = housing[housing['Thumbnail Url'].notnull()]
print('Nº nulos ahora:', housinNoNulaUrl["Thumbnail Url"].isnull().values.sum())
```

## Descarga de imágenes
Inicio de la descarga de imágenes

```python
# Crear nueva columna para guardar la ruta de la imagen descargada
import numpy as np

housinNoNulaUrl["imagenLocal"] = np.nan
housinNoNulaUrl.head()
```

```python
import urllib.request
from os.path import join

def downloadImage(pIdElemento,pUrl):        
    rutaImagen = 'img'

    try:
        filename = '{}.jpg'.format(pIdElemento)    
        full_path = join(rutaImagen, filename)
        
        urllib.request.urlretrieve(pUrl, full_path)

        return full_path
    except Exception as e:
        return np.nan
```

```python
for index, row in housinNoNulaUrl.iterrows():
    rutaImagenLocal = downloadImage(row["ID"], row["Thumbnail Url"])    
    
    housinNoNulaUrl.loc[index,"imagenLocal"] = rutaImagenLocal
```

Se comprueba las imágenes que no han podido descargarse

```python
print('Nº nulos imagen:', housinNoNulaUrl["imagenLocal"].isnull().values.sum())
```

Eliminamos esos registros

```python
housingNoNulaUrl = housinNoNulaUrl[housinNoNulaUrl['imagenLocal'].notnull()]
print('Nº nulos ahora:', housingNoNulaUrl["imagenLocal"].isnull().values.sum())
```

Guardo el nuevo dataframe en un nuevo csv

```python
housingNoNulaUrl.to_csv('./datasets/datasetImagenes.csv', sep=';', decimal='.', index=False)
```

```python
housing = load_airbnb_dataset('datasets','datasetImagenes.csv')
housing.head()
```

## Observar el tamaño de las imágenes

```python
from PIL import Image

imagenesOtros = []

for url in housing['imagenLocal']:
    width, height = Image.open(url).size

    if (width != 216) or (height != 144):
        imagenesOtros.append(url)

imagenesOtros
```

Hay 7 imágenes con tamaños diferentes a los establecidos, al ser tan pocas no se realiza el resize, directamente borramos esas filas del dataset

```python
housing.shape
```

```python
nuevoHousing = housing[housing['imagenLocal'].apply(lambda x: x not in imagenesOtros)]
nuevoHousing.shape
```

## Guardar dataset final
Se guarda el dataset que contiene imágenes para todas las filas como paso final.

```python
nuevoHousing.to_csv('./datasets/datasetImagenesFinal.csv', sep=';', decimal='.', index=False)
```
